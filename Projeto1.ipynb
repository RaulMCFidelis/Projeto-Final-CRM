```python
# 1. Tudo minúsculo
df_rmx['dia_semana'] = df_rmx['dia_semana'].str.lower()

# 2. Remove espaços em branco no início ou no fim
df_rmx['dia_semana'] = df_rmx['dia_semana'].str.strip()

# 3. Remove a palavra "-feira"
df_rmx['dia_semana'] = df_rmx['dia_semana'].str.replace('-feira', '', regex=False)

# 4. Tira acentos e cedilha (terça -> terca, sábado -> sabado)
df_rmx['dia_semana'] = df_rmx['dia_semana'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8')
```

```python
# Criamos o "dicionário tradutor"
mapa_dias = {
    'segunda': 0,
    'terca': 1,
    'quarta': 2,
    'quinta': 3,
    'sexta': 4,
    'sabado': 5,
    'domingo': 6
}

# Criamos uma nova coluna numérica baseada na coluna de texto
df_rmx['dia_semana_num'] = df_rmx['dia_semana'].map(mapa_dias)
```

```python
# Cria uma coluna que recebe 1 se for Sábado(5) ou Domingo(6), e 0 se for dia útil
df_rmx['is_fim_de_semana'] = df_rmx['dia_semana_num'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
df_rmx = df_rmx.drop(columns=['dia_semana'])
```

```python
# Dicionário com a ordem lógica das idades
mapa_idade = {
    '18-24': 0,
    '25-34': 1,
    '35-44': 2,
    '45-54': 3,
    '55+': 4
}

# Nova coluna numérica
df_rmx['faixa_etaria_num'] = df_rmx['faixa_etaria'].map(mapa_idade)

# Dropei a coluna de texto antiga para limpar a base
df_rmx.drop(columns=['faixa_etaria'], inplace=True)
```
